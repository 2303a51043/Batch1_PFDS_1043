```python
import numpy as np
import matplotlib.pyplot as plt

monthly_temps_c = np.array([5.2, 7.5, 12.0, 16.4, 20.1, 24.5, 27.0, 26.5, 22.0, 16.8, 10.2, 6.1])
avg_temp = np.mean(monthly_temps_c)
hottest_month = np.argmax(monthly_temps_c) + 1
coldest_month = np.argmin(monthly_temps_c) + 1
normalized_temps = (monthly_temps_c - monthly_temps_c.min()) / (monthly_temps_c.max() - monthly_temps_c.min())
above_avg_months = np.where(monthly_temps_c > avg_temp)[0] + 1
temps_fahrenheit = monthly_temps_c * 9/5 + 32
plt.plot(range(1, 13), monthly_temps_c, marker='o', label='Celsius')
plt.plot(range(1, 13), temps_fahrenheit, marker='x', label='Fahrenheit', linestyle='--')
plt.title("Monthly Temperature")
plt.xlabel("Month")
plt.ylabel("Temperature")
plt.legend()
plt.grid(True)
plt.show()
print("Monthly Temperatures (°C):", monthly_temps_c)
print("Average Temperature (°C):", round(avg_temp, 2))
print("Hottest Month:", hottest_month)
print("Coldest Month:", coldest_month)
print("Normalized Temperatures:", np.round(normalized_temps, 2))
print("Above Average Months:", above_avg_months)
print("Temperatures in Fahrenheit (°F):", np.round(temps_fahrenheit, 2))
```

```python
import pandas as pd
data = {
    'Order ID': [101, 102, 103, 104, 105, 106, 107],
    'Product': ['Laptop', 'Phone', 'Tablet', 'Laptop', 'Headphones', 'Phone', 'Tablet'],
    'Quantity': [1, 2, 3, 1, 5, 2, 4],
    'Price': [1200, 600, 300, 1300, 100, 650, 280]
}
df = pd.DataFrame(data)
df['Total Sale'] = df['Quantity'] * df['Price']
high_value_orders = df[df['Total Sale'] > 1000]
product_sales = df.groupby('Product')['Quantity'].sum()
most_sold_product = product_sales.idxmax()
df['Discount'] = df['Total Sale'].apply(lambda x: x * 0.10 if x > 1000 else x * 0.05)
df.to_excel('ecommerce_sales_summary.xlsx', index=False)
print("Full DataFrame:")
print(df)
print("\nHigh Value Orders (> $1000):")
print(high_value_orders)
print("\nMost Sold Product:", most_sold_product)
```
